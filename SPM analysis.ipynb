```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Sigmoid Function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))


x1 = np.arange(-5.0, 5.0, 0.1)
y1 = sigmoid(x1)

def sigmoid2(x):
    return 1 / (1 + np.exp(x))
x2 = np.arange(-5.0, 5.0, 0.1)
y2 = sigmoid2(x2)

plt.plot(x1, y1,'r')
plt.plot(x2, y2,'b')
plt.ylim(-0.1, 1.1)
plt.show()
```
