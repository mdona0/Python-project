```python
#f(x) = x
def diff(f,x):
    h = 0.0001
    return (f(x+h)-f(x))/h

def f1(x):
    return x

x = [0.0001*i for i in range(1,100000)]
y = [f1(i) for i in x]
dfdx = [diff(f1, i) for i in x]
plt.plot(x,y,'r')
plt.plot(x, dfdx,'b')
plt.show()
```

```python
# f(x) = x^2 + x +1
def f2(x):
    return x**2 + x + 1

x = [0.0001*i for i in range(1,100000)]
y = [f2(i) for i in x]
dfdx = [diff(f2, i) for i in x]
plt.plot(x,y,'r')
plt.plot(x, dfdx,'b')
plt.show()
```

```python
# f(x) = exp(x)
def fe(x):
    return math.e**x

x = [0.0001*i for i in range(1,100000)]
y = [fe(i) for i in x]
dfdx = [diff(fe, i) for i in x]
plt.plot(x,y,'r')
plt.plot(x, dfdx,'b')
plt.show()
```

```python
#many-valued function
def fx(x):
    y = 1
    return x**2+2*y**2

def fy(y):
    x = 1
    return x**2+2*y**2

def diff(f,p):
    h = 0.0001
    return (f(p+h)-f(p))/h

x = [0.0001*i for i in range(1,100000)]
y = [0.0001*i for i in range(1,100000)]

#∂x
delxf = [diff(fx, i) for i in x]

#∂y
delyf = [diff(fy, i) for i in y]

plt.plot(x,y,'r')
plt.plot(x, delxf,'b')
plt.show()

plt.plot(x,y,'r')
plt.plot(x, delyf,'b')
plt.show()
```

```python
# definite integral
def f(x):
    return x**2

x = [i for i in range(0,100)]
a = 20
b = 80
y = [f(i) for i in x]
plt.plot(x, y,'b')
plt.axvline(x=a, c='b') #x=a
plt.axvline(x=b, c='b') #x=b
#plt.axvline(y=0, c='b') #x of axis
plt.show()

dy = [(f(b)-f(a))*(i-a)/(b-a)+f(a) for i in x]
plt.plot(x, dy, 'r')
```

```python
#part2
def f(x):
    return x**2

def integral(f, a, b, n):
    c = np.linspace(a, b, n)
    ans = 0
    for i in range(n-1):
        ans += f(c[i]+f(c[i+1]))
    return (b-a)*ans/(2*n)

a = 20
b = 80
n = 100
integral(f,a,b,n)
```

```python
def Mysimulation(f, a, b, I):
    N = [i for i in range(20,10000,10)]
    y = [integral(f,a,b,n)-I for n in N]
    plt.scatter(N, y)
    
I = 168000
a = 20
b = 80
Mysimulation(f, a, b, I)
```
