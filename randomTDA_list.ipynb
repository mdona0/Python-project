```python
import numpy as np
from ripser import ripser
from persim import plot_diagrams
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as stats
```

```python
#正規分布に従うランダムな2次元データセットの場合
```

```python
data = np.random.random((100, 2))
data = pd.DataFrame(data)
plt.title('μ=0, σ^2=1,normal distribution')
plt.scatter(data[0], data[1])
```

```python
#大数の法則、平均収束
n = 1000
mean  = 0.0
mean_his = np.zeros(n)
count = 0


num_list = []

for i in range(n):
    count += 1
    data = np.random.random((100, 2))
    
    diagrams = ripser(data)['dgms']

    num_list.append(len(diagrams[1]))
    
    mean += (len(diagrams[1]) - mean) / count
    mean_his[i] = mean
    
plt.scatter(x=range(n), y=mean_his)
plt.title('convergence of betti number')
plt.show()    
```

```python
#ヒストグラム
plt.hist(num_list)
plt.title('histgram of betti number')
plt.show()
```

```python
#Q-Qプロット
fig = plt.subplots()
 
stats.probplot(num_list, dist="norm", plot=plt)
plt.show()
```

```python
#シャピロウィルク検定
stats.shapiro(num_list, "norm")
```
