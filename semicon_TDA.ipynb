```python
import matplotlib.pyplot as plt
import glob
import numpy as np
from ripser import ripser
from persim import plot_diagrams

import warnings
warnings.simplefilter('ignore')

DIF1 = glob.glob('./sck_large1,2/Large2/DIF1/OK/*.png')

DIF1_list = []

for i in DIF1:
    img = plt.imread(i)
    DIF1_list.append(img)
```

```python
data = DIF1_list[0]
diagrams = ripser(data)['dgms']
print(len(diagrams[1]))
plot_diagrams(diagrams, show=True)
```

```python
num_list = []

for j in DIF1_list:
    diagrams = ripser(j)['dgms']
    num_list.append(len(diagrams[1]))

#plt.hist(num_list)
plt.show()

print(np.mean(num_list))
print(np.var(num_list))
```
